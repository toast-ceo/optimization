```python
import numpy as np
import scipy.io
import matplotlib.pyplot as plt
```

```python
# MAT 파일 불러오기
mat_path = r"../HW2_package/hw2_prob2.mat"
mat_data = scipy.io.loadmat(mat_path)

# 변수 할당
A = mat_data['A']
d = mat_data['d']
d_up = mat_data['d_up']
```

```python
# minimum f
def f(x):
    return 0.5 * np.linalg.norm((A @ x - d), 2) 
    
def f_1_differential(x):
    return A.T @ (A @ x - d)
```

```python
# logarithmic_barrier
def logarithmic_barrier(x):
    return - np.log(d_up - (A @ x))

def logarithmic_barrier_1_differential(x):
    return A.T @ np.reciprocal(d_up - (A @ x))
```

```python
def F(x, t):
    return t * f(x) + logarithmic_barrier(x)
    
def F_1_differential(x, t):
    return t * f_1_differential(x) + logarithmic_barrier_1_differential(x) 
```

```python
#결과 저장
arr_F = []
arr_F_xk = []
arr_f = []
arr_f_xk = []
```

```python
# without inequality constraint

t = 2
xk = np.zeros((A.shape[1], 1))

# 조정 값
alpha = 0.01
beta = 0.8



for idx in range(10000):
    
    delta_x = - f_1_differential(xk)
    step_size = 1

    #feasibility check
    while np.max(A @ (xk + t * delta_x)) >= 1 or np.max(xk + t * delta_x) >= 1:
        t = beta * t

    #backtracking line search
    while f(xk + step_size * delta_x) > (f(xk) + alpha * step_size * f_1_differential(xk).T @ delta_x):
        step_size *= beta
    
    xk = xk + step_size * delta_x
    
    
    print(idx, xk, f(xk))
    arr_f_xk.append(f(xk))
    arr_f_xk.append(xk)
```
