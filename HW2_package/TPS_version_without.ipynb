```python
import numpy as np
import cupy as cp
import scipy.io
import matplotlib.pyplot as plt
```

```python
# MAT 파일 불러오기
mat_path = r"../HW2_package/hw2_prob2.mat"
mat_data = scipy.io.loadmat(mat_path)

# 변수 할당
A = cp.array(mat_data['A'])
d = cp.array(mat_data['d'])
d_up = cp.array(mat_data['d_up'])
t = 2
```

```python
def f(x):
    return 0.5 * (cp.linalg.norm(cp.matmul(A, x) - d, ord=2) ** 2)

def f_1_differential(x):
    return A.T @ (A @ x - d)
    
def f_1_differential_H(x, temp_d):
    AX = A @ x 
    mask = AX > temp_d
    AX = AX * (1 - mask)  # temp_d 보다 큰 값은 0으로 설정
    return A.T @ (AX - d)
```

```python
def get_DVH_PTV(PTV, x_beam):
    x_ax = cp.arange(0, 70.1, 0.1)
    PTV_dummy = PTV @ x_beam.flatten()

    Dummy_t, _ = cp.histogram(PTV_dummy, bins=x_ax)
    DVH_t = cp.flip(cp.cumsum(cp.flip(Dummy_t)))
    DVH_PTV = DVH_t / cp.max(DVH_t)

    return DVH_PTV
```

```python

# 변수 초기화
x_H = cp.zeros((A.shape[1], 1))

# 조정 값 F
alpha = 0.01
beta = 0.8

# 반복 횟수
iterations = 1000

for idx in range(iterations):
    delta_x = -f_1_differential_H(x_H, d_up)
    step_size = 1

    # backtracking line search
    while f(x_H + step_size * delta_x) > (f(x_H) + alpha * step_size * f_1_differential(x_H).T @ delta_x):
        step_size *= beta

    x_H = x_H + step_size * delta_x
    print(idx)
    
print("done")
```

```python
# 변수 초기화
x_nH = cp.zeros((A.shape[1], 1))

# 조정 값 F
alpha = 0.01
beta = 0.8

iterations = 1000

for idx in range(iterations):
    delta_x = -f_1_differential(x_nH)
    step_size = 1

    # backtracking line search
    while f(x_nH + step_size * delta_x) > (f(x_nH) + alpha * step_size * f_1_differential(x_nH).T @ delta_x):
        step_size *= beta
    
    x_nH = x_nH + step_size * delta_x
    print(idx)

print("done")
```

```python
DVH_H = get_DVH_PTV(A, x_H)
DVH_nH = get_DVH_PTV(A, x_nH)
```
